```python
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
import itertools

pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 100)
pd.set_option('display.width', 100)
```

```python
file_name = 'megasena.xlsx' 
df = pd.read_excel(file_name, index_col=0)
df = df.reset_index()
df.head(10)
```

```python
df2 = df.drop(['As Loterias - www.asloterias.com.br - Todos Resultados da Mega Sena'], axis=1)
df2.columns = ['data', 'n1', 'n2', 'n3', 'n4', 'n5', 'n6']
df2 = df2.dropna()
df2 = df2.drop(5)
df2 = df2.set_index('data') 
df2.head(5)
```

```python
def numconv(df):
    lista = df.values.tolist()
    lista = list(itertools.chain(*lista))
    df = DataFrame (lista, columns=['numbers'])
    df = df['numbers'].value_counts()
    return df
```

```python
def filtro(df,a,b,c,d,e):
    searchfor = [str(a), str(b), str(c), str(d), str(e)]
    df = df.reset_index()
    df = df[df.data.str.contains('|'.join(searchfor))]
    df = df.set_index('data') 
    return df
```

```python
dftotal = numconv(df2)
dftotal.head()
```

```python
plt.figure(figsize=(15,5))
ax = plt.gca()                                                      
plt.title('Recorrência dos números da Mega Sena 1996-2020')             

dftotal.plot(kind='bar',x= 'number', y='vezes', ax=ax)

plt.ylabel('repetições')                                              
plt.xlabel('número')                                             
plt.xticks(rotation=45)                                            
ax.set_axisbelow(True)                                                 
ax.yaxis.grid(color='gray', linestyle='dashed')
```

```python
df2020 = filtro(df2, 2020, 2019, 2018, 2017, 2016)
df2020 = numconv(df2020)
df2020.head()
```

```python
plt.figure(figsize=(15,5))
ax = plt.gca()                                                      
plt.title('Recorrência dos números da Mega Sena 2015-2020')             

df2020.plot(kind='bar',x= 'number', y='vezes', ax=ax)

plt.ylabel('repetições')                                              
plt.xlabel('número')                                             
plt.xticks(rotation=45)                                            
ax.set_axisbelow(True)                                                 
ax.yaxis.grid(color='gray', linestyle='dashed')
```
